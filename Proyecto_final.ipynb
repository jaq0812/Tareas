# **Análisis de Créditos Infonavit Formalizados (Enero - Febrero 2022)**



*- Introducción*

El proposito de este informe es analizar la distribución y las tendencias de los creditos infonavit formalizados durante los meses de enero y febrero de 2022, ya que en este año hubieron ajustes significativos en el mercado de crédito para vivienda en mexico, con ello, se presento una descancelación en la colocación de créditos, una disminución en la participación de mercado de infonavot y un incremento en la cartera vencida. Sin embargo, Infonavit también implementó medidas como el programa Crediterreno y un aumento en la edad máxima para los solicitantes. Este análisis busca proveer una perspectiva microeconómica sobre cómo estos factores macroeconómicos y las políticas institucionales impactaron la formalización de créditos a nivel demográfico y de producto.


Se examinará cómo se distribuyeron estos créditos entre diferentes grupos de edad, tipos de producto y estados de México, con el fin de identificar patrones clave y comprender mejor la dinámica del mercado crediticio para vivienda en un periodo de desafíos económicos y cambios institucionales.

*- Descripción de los datos*

El conjunto de datos utilizado para este análisis, denominado "Masivos_IOC_2025_6_4_21_35_35.txt", contiene información detallada sobre los créditos Infonavit formalizados en enero y febrero de 2022. Las columnas principales del dataset son:

Modalidad de participación: Describe el tipo de participación en el crédito .

Tipo de producto: Especifica el tipo de crédito Infonavit otorgado.

Estados: Indica la entidad federativa de México donde se formalizó el crédito.

Edad: Clasifica a los beneficiarios por rangos de edad.

Número de créditos formalizados: Representa el recuento de créditos formalizados para cada combinación única de las variables anteriores.

Este conjunto de datos permite un análisis detallado de la colocación de créditos por diversas dimensiones, facilitando la identificación de segmentos demográficos y de producto clave.

*- Metodología*

Para llevar a cabo este análisis, se empleó un enfoque basado en estadísticas descriptivas y análisis comparativo. La metodología seguida comprende los siguientes pasos:

1. Carga y Preprocesamiento de Datos: El archivo de texto fue cargado en un DataFrame de pandas. Se realizó una limpieza inicial para asegurar la correcta interpretación de los datos, incluyendo la conversión de la columna Número de créditos formalizados a un formato numérico y la estandarización de los nombres de las columnas. Además, se estableció un orden lógico para las categorías de edad para una mejor visualización.

2. Agregación de Datos:

  * Se calculó el número total de créditos formalizados por Grupo de Edad.

  * Se determinó el número total de créditos formalizados por Tipo de Producto.

  * Se sumaron los créditos formalizados por Estado para identificar las regiones con mayor actividad crediticia.

3. Tabulación Cruzada: Se generó una tabla de contingencia para analizar la relación entre Edad y Tipo de producto, mostrando la cantidad de créditos formalizados para cada combinación.

4. Visualización de Datos: Se crearon gráficos de barras utilizando matplotlib y seaborn para representar visualmente las distribuciones de créditos por grupo de edad, tipo de producto y los 10 estados principales.

5. Interpretación y Discusión: Los resultados numéricos y las visualizaciones fueron analizados para identificar patrones, tendencias y disparidades en la colocación de créditos, lo cual permitió una discusión crítica en el contexto de las condiciones del mercado hipotecario mexicano en 2022.

*- Resultados*

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'Masivos_IOC_2025_6_4_21_35_35.txt'
df = pd.read_csv(file_path, sep='|')

df.columns = df.columns.str.strip()

df = df.rename(columns={
    'Modalidad de participación': 'Modalidad_participacion',
    'Tipo de producto': 'Tipo_producto',
    'Estados': 'Estado',
    'Edad': 'Edad',
    'Número de créditos formalizados': 'Creditos_formalizados'
})

# Convertimos 'Creditos_formalizados' a tipo numerico
df['Creditos_formalizados'] = pd.to_numeric(df['Creditos_formalizados'], errors='coerce').fillna(0)


creditos_por_edad = df.groupby('Edad', observed=True)['Creditos_formalizados'].sum()
print("Tabla de créditos formalizados por grupo de edad:\n")
print(creditos_por_edad)

plt.figure(figsize=(12, 6))
sns.barplot(x=creditos_por_edad.index, y=creditos_por_edad.values, palette='viridis')
plt.title('Créditos formalizados por grupo de edad (Ene-Feb 2022)')
plt.xlabel('Grupo de edad')
plt.ylabel('Número de créditos formalizados')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()


print("\n --- Créditos formalizados por tipo de producto --- \n")

creditos_por_producto = df.groupby('Tipo_producto')['Creditos_formalizados'].sum().sort_values(ascending=False)
print("Tabla de créditos formalizados por tipo de producto:\n")
print(creditos_por_producto)

plt.figure(figsize=(12, 6))
sns.barplot(x=creditos_por_producto.index, y=creditos_por_producto.values, palette='magma')
plt.title('Créditos formalizados por tipo de producto (Ene-Feb 2022)')
plt.xlabel('Tipo de producto')
plt.ylabel('Número de créditos formalizados')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()


print("\n --- Créditos Formalizados por Estado (Top 10) --- \n")

creditos_por_estado = df.groupby('Estado')['Creditos_formalizados'].sum().sort_values(ascending=False)
top_10_estados = creditos_por_estado.head(10)
print("Tabla de créditos formalizados por estado (Top 10):\n")
print(top_10_estados)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_10_estados.index, y=top_10_estados.values, palette='cividis')
plt.title('Top 10 estados con mayor número de créditos formalizados (Ene-Feb 2022)')
plt.xlabel('Estado')
plt.ylabel('Número de créditos formalizados')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()


print("\n --- Tabulación cruzada: Créditos formalizados por edad y tipo de producto --- \n")

crosstab_edad_producto = pd.crosstab(df['Edad'], df['Tipo_producto'], values=df['Creditos_formalizados'], aggfunc='sum').fillna(0).astype(int)
print("Tabla de tabulación cruzada (Créditos formalizados por edad y tipo de producto):\n")
print(crosstab_edad_producto)
```

*- Conclusiones*

En base a los resultados podemos concluir que:

La distribución de créditos por grupo de edad muestra una concentración en rangos intermedios. Los grupos de 26 a 30 años y 31 a 35 años son los que formalizaron la mayor cantidad de créditos, seguidos por el grupo de 36 a 40 años. La participación es menor en las edades extremas, es decir, en los solicitantes muy jóvenes (hasta 20 años) y en los mayores de 50 años.


El analisis por tipo de producto destaca el predominio del "Crédito Tradicional", que concentra la mayoría de las formalizaciones. Otros productos como "Mejoravit" y "Cofinavit" también contribuyen, pero en una escala menor. Los productos "Infonavit Total", "Apoyo Infonavit" y "Tu segundo crédito" representan una porción marginal del total de créditos.


Estado de México y Nuevo León lideran la formalización de créditos Infonavit durante este periodo, lo que sugiere una alta demanda o una robusta actividad inmobiliaria en estas regiones. Jalisco y la Ciudad de México también muestran un volumen significativo de créditos, haciendo a estas áreas como centros clave de la actividad crediticia de Infonavit.


El "Crédito Tradicional" es el producto principal en casi todos los grupos de edad, siendo especialmente para entre los 26 y 45 años. "Mejoravit" y "Cofinavit" también muestran actividad en estos mismos rangos de edad, lo que indica que son opciones populares para mejoras de vivienda o cofinanciamiento entre la población. Es notable que la categoría "Otros" en edad para "Apoyo Infonavit" y "Mejoravit" sugiere flexibilidad en estos productos para segmentos de edad no estandarizados o casos especiales.
